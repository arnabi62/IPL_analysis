```python
wicket_per_ball = wicket/ball
print(wicket_per_ball)
wkt_max = max(wicket_per_ball)
eco_max = max(eco)
sr_max = max(sr)
wkt_max

c_wicket_per_ball = c_wicket/c_ball
print(c_wicket_per_ball)
c_wkt_max = max(c_wicket_per_ball)
c_eco_max = max(c_eco)
c_sr_max = max(c_sr)
c_wkt_max
```

```python
wicket_cost = []
for i in range(len(player)):
    wicket_cost.append(wicket_per_ball[i]/wkt_max)
print(wicket_cost)

c_wicket_cost = []
for i in range(len(player)):
    c_wicket_cost.append(c_wicket_per_ball[i]/c_wkt_max)
c_wicket_cost
```

```python
max_ave = max(ave)
ave_cost = []
for i in range(len(player)):
    ave_cost.append(ave[i]/max_ave)
print(ave_cost)

c_max_ave = max(c_ave)
c_ave_cost = []
for i in range(len(player)):
    c_ave_cost.append((c_ave[i]/c_max_ave))
#c_ave_cost
```

```python
eco_cost = []
for i in range(len(player)):
    eco_cost.append((eco[i]/eco_max))
eco_cost

c_eco_cost = []
for i in range(len(player)):
    c_eco_cost.append((c_eco[i]/c_eco_max))
c_eco_cost
```

```python
sr_cost = []
for i in range(len(player)):
    sr_cost.append((sr[i]/sr_max))
sr_cost


c_sr_cost = []
for i in range(len(player)):
    c_sr_cost.append((c_sr[i]/c_sr_max))
c_sr_cost
```

```python
strike_bowl = []
for i in range(len(player)):
    big = (4* four[i]+ 5*five[i])/(ball[i]*6)*(wicket_cost[i])
    strike_bowl.append(big*100)
strike_bowl

```

```python
c_strike_bowl = []
for i in range(len(player)):
    c_big = (4* c_four[i]+ 5*c_five[i])/(c_ball[i]*6)
    c_strike_bowl.append(c_big*100)
c_strike_bowl
```

```python
mean_inn = inn.mean()
#print(mean_inn)
x_fact = []
for i in range(0, len(player)):
        exp = inn[i]/mean_inn
        x_fact.append(exp)      
    


act_x_fact = []
m_x_fact = max(x_fact)
for i in range(len(player)):
    act_x_fact.append(x_fact[i]/m_x_fact)
print(act_x_fact)
std = (mean(act_x_fact)-min(act_x_fact))

bin = [0,0.3,0.6,0.9,1.2]
plt.hist(act_x_fact,bin, histtype = 'bar',rwidth=0.8)
std_x_fact = []
for i in range(len(player)):
    if(act_x_fact[i] > std):
        std_x_fact.append(1)
    else:
        std_x_fact.append(0.5)
    
std_x_fact
```

```python
handy_bowl = []
for i in range(len(player)):
    perform = (wicket[i] - 4 * four[i] - 5 * five[i])/(ball[i] - four[i] - five[i])
    handy_bowl.append(perform)
print(handy_bowl)

c_handy_bowl = []
for i in range(len(player)):
    c_perform = (c_wicket[i] - 4 * c_four[i] - 5 * c_five[i])/(c_ball[i] - c_four[i] - c_five[i])
    c_handy_bowl.append(c_perform)
#c_handy_bowl
```

```python
cr=np.array(final_point)
crmean=cr.mean()
crmeanup=[]
crmeanless=[]
for i in range(0,len(player)):
    if(final_point[i]>=crmean):
        crmeanup.append(final_point[i])
    else:
        crmeanless.append(final_point[i])
crup=np.array(crmeanup)
crless=np.array(crmeanless)
crupmean=crup.mean()
crlessmean=crless.mean()
print(crupmean)
print(crlessmean)
```
